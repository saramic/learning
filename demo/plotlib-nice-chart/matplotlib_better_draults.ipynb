````python
# import sys
# !{sys.executable} -m pip install matplotlib

import matplotlib.pyplot as plt
import matplotlib as mpl

# reset to defaults
mpl.rcParams.update(mpl.rcParamsDefault)

# ── Figure size & DPI ────────────────────────────────────────────────────────
mpl.rcParams["figure.figsize"]   = (12, 7)    # wide landscape; tweak to taste
mpl.rcParams["figure.dpi"]       = 150        # crisp on retina / HD screens
mpl.rcParams["savefig.dpi"]      = 300        # publication-quality on export
mpl.rcParams["savefig.bbox"]     = "tight"    # never clip labels on save

# ── Fonts ─────────────────────────────────────────────────────────────────────
mpl.rcParams["font.family"]      = "sans-serif"
mpl.rcParams["font.sans-serif"]  = ["DejaVu Sans"]  # always available; swap
                                                      # for "Helvetica" etc.
mpl.rcParams["font.size"]        = 13
mpl.rcParams["axes.titlesize"]   = 16
mpl.rcParams["axes.labelsize"]   = 14
mpl.rcParams["xtick.labelsize"]  = 12
mpl.rcParams["ytick.labelsize"]  = 12
mpl.rcParams["legend.fontsize"]  = 12

# ── Lines & markers ───────────────────────────────────────────────────────────
mpl.rcParams["lines.linewidth"]  = 2.0
mpl.rcParams["lines.markersize"] = 7

# ── Axes ──────────────────────────────────────────────────────────────────────
mpl.rcParams["axes.spines.top"]    = False   # remove top & right frame lines
mpl.rcParams["axes.spines.right"]  = False
mpl.rcParams["axes.grid"]          = True
mpl.rcParams["axes.axisbelow"]     = True    # grid behind data, not on top
mpl.rcParams["axes.titlepad"]      = 12      # breathing room above title

# ── Grid ──────────────────────────────────────────────────────────────────────
mpl.rcParams["grid.color"]         = "#cccccc"
mpl.rcParams["grid.linestyle"]     = "--"
mpl.rcParams["grid.linewidth"]     = 0.7
mpl.rcParams["grid.alpha"]         = 0.7

# ── Color cycle (10-color Tableau palette — colorblind-friendlier) ─────────
mpl.rcParams["axes.prop_cycle"] = mpl.cycler(
    color=[
        "#4C72B0", "#DD8452", "#55A868", "#C44E52", "#8172B3",
        "#937860", "#DA8BC3", "#8C8C8C", "#CCB974", "#64B5CD",
    ]
)

# ── Layout ────────────────────────────────────────────────────────────────────
mpl.rcParams["figure.constrained_layout.use"] = True   # auto-fix overlapping
                                                       # labels / subplots

# ── Legend ────────────────────────────────────────────────────────────────────
mpl.rcParams["legend.framealpha"]  = 0.9
mpl.rcParams["legend.edgecolor"]   = "#cccccc"
mpl.rcParams["legend.borderpad"]   = 0.6

# ─────────────────────────────────────────────────────────────────────────────
# ALTERNATIVE: load as a named style instead of setting rcParams inline.
# Save all the above as  ~/.config/matplotlib/stylelib/clean.mplstyle
# then call:  plt.style.use("clean")
#
# Or use one of matplotlib's built-in styles as a starting point:
#   plt.style.use("seaborn-v0_8-whitegrid")   # clean white + grid
#   plt.style.use("ggplot")                    # R ggplot look
#   plt.style.use("bmh")                       # Bayesian Methods for Hackers
# ─────────────────────────────────────────────────────────────────────────────


import numpy as np

plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.title("Trig Functions")
plt.xlabel("x (radians)")
plt.ylabel("Amplitude")
plt.legend()


plt.show()
````

````python
x = np.linspace(0, 2 * np.pi, 200)

fig, axes = plt.subplots(1, 2)

axes[0].plot(x, np.sin(x), label="sin(x)")
axes[0].plot(x, np.cos(x), label="cos(x)")
axes[0].set_title("Trig Functions")
axes[0].set_xlabel("x (radians)")
axes[0].set_ylabel("Amplitude")
axes[0].legend()

rng = np.random.default_rng(42)
axes[1].scatter(rng.normal(size=120), rng.normal(size=120),
                alpha=0.7, label="samples")
axes[1].set_title("Scatter Plot")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")
axes[1].legend()

plt.savefig("demo.png")
plt.show()
````
