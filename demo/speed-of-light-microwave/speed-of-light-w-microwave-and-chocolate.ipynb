# Measuring the Speed of Light with a Microwave Oven

## Concept

A microwave oven cavity supports a **standing electromagnetic wave** at 2.45 GHz.
Standing waves have fixed points of maximum field intensity (antinodes) spaced
**λ/2** apart. By melting a flat sheet of chocolate at these hot spots and
measuring the distance between them, we can back out the wavelength — and from
there, the speed of light.

**Governing equation:**

$$c = f \times \lambda$$

Where:
- `f` = magnetron frequency, printed inside the oven or on the rating plate (almost always **2450 MHz**)
- `λ` = wavelength, calculated from measured hot-spot spacing

**Key relationship:** adjacent melt spots are **λ/2** apart, not λ. Always double
your measured spot-to-spot distance before plugging into the equation above —
this is the single most common source of a 2x error in this experiment.

---

## Materials

| Item | Notes |
|---|---|
| Milk chocolate (thin, flat — e.g. Lindt-style bar) | Sharp, low melting point (~30–32°C); thin cross-section avoids lateral heat smearing |
| Microwave oven | Turntable **removed** — static field pattern is essential |
| Flat heatproof plate or tray | Large enough to cover most of the cavity floor |
| Ruler / calipers | For measuring hot-spot spacing |
| (Optional) TinySA / TinySA Ultra | For a leakage-detection side experiment |

### Chocolate choice

- **Lindt-style thin flat bar → best.** Uniform thin cross-section melts near-simultaneously through its depth, giving clean, distinct dots.
- **Cadbury thick block → worse.** More thermal mass and longer heating time let heat spread sideways before spots are distinguishable — smeared blobs instead of dots.
- **Toblerone → don't bother.** The triangular ridge spacing is a property of the mold, unrelated to the λ/2 ≈ 6.1 cm field spacing at 2.45 GHz. Peaks/troughs will not correlate with the RF standing wave pattern — just adds a confounding variable.

---

## Procedure

1. **Remove the turntable** — lift out the glass tray and the roller ring/support. This is critical; a rotating plate will smear the hot spots into an unusable blur.
2. **Isolate the plate from the drive nub.** Place an upside-down heatproof bowl, or a flat non-metallic spacer (book, silicone mat) over the central drive coupling on the cavity floor, then set your chocolate plate on top so it can't catch on the rotating hub.
3. **Quick spin test:** run the empty microwave a few seconds and confirm nothing on the floor rotates.
4. **Lay the chocolate flat**, covering as much of the cavity floor as practical.
5. **Heat in short bursts** — start with 15–20 seconds, check, repeat until you see 2–4 distinct melt spots (not full melt).
6. **Measure spacing** between adjacent spots. For better accuracy, measure across several spots (e.g. spot 1 → spot 4 = 3 gaps) and divide by the number of gaps to average out measurement error.
7. **Double the measured spacing** to get λ (since spots are λ/2 apart).
8. **Calculate:**

$$c = f \times \lambda$$

With a typical measured λ ≈ 12.2 cm and f = 2450 MHz:

$$c \approx 2.45\times10^{9} \times 0.122 \approx 2.99\times10^{8}\ \text{m/s}$$

— within a couple percent of the accepted value.

## Optional: RF Leakage Check with a TinySA

A nice complementary measurement: check whether the oven leaks RF at the door seal.

### Safety first — no internal protection to rely on
- Cheap spectrum analyzers (TinySA / TinySA Ultra) typically have only minimal diode-clamp protection, rated for transient ESD, **not sustained high power**.
- A magnetron runs at ~700–1100 W inside the cavity; even a small leaking fraction can exceed the analyzer's safe input (~0 to +10 dBm continuous, model-dependent).
- **No attenuator on hand → use distance instead:**
  1. Start several meters away, oven running, analyzer in max-hold.
  2. Move closer in ~1 m increments, watching the trace each time.
  3. Stop and back off immediately if you see a flat-topped/clipped peak, odd spurs, or the reading failing to keep rising as you approach (signs of front-end saturation).
  4. Never bring the probe within a few cm of the door seam/mesh, even if readings look low from a distance — near-field leakage can be concentrated.

### Instrument setup
- **RBW:** Auto is fine, or widen to 100–300 kHz for faster sweeps and better chance of catching a transient leak — no need for fine resolution here.
- **Trace mode: Max Hold** — leakage is intermittent as the magnetron cycles, so a single sweep can miss it.
- **Center frequency:** 2450 MHz, narrow span or zero-span.
- On the device: `Settings → RBW` and `Display/Trace → Max Hold` (menu names vary slightly by firmware).

### What to look for
- Scan around the door edges, mesh window, and seams.
- A well-sealed oven: near-nothing above the ambient 2.4 GHz noise floor (Wi-Fi/BT background is normal and will be present regardless).
- A worn seal/door: a distinct spike that scales with proximity to the oven and vanishes when the oven is switched off — that's your leakage signature, distinguishable from ambient Wi-Fi/BT traffic.

## Summary of Key Numbers

| Quantity | Value |
|---|---|
| Magnetron frequency (f) | 2450 MHz (check oven's rating plate) |
| Expected wavelength (λ) | ≈ 12.2 cm |
| Expected hot-spot spacing (λ/2) | ≈ 6.1 cm |
| Melting chocolate | Milk, thin & flat (Lindt-style) |
| Burst heating time | 15–20 s, check and repeat |
| Calculated c | ≈ 2.99 × 10⁸ m/s (accepted: 2.998 × 10⁸ m/s) |
