# Precission

Looking at approximating **log-scale magnitude as integer precision**. Small differences near powers of 10 make that inherently fuzzy.

If $\text{error} \approx 10^{-d}$, then you have $\sim d$ digits of precision

Example boundaries:

Error  | log10 | Digits
-------|-------|-------
0.0099 | -2.00 | ~2
0.0011 | -2.96 | ~3
0.0010 | -3.00 | 3 exactly

my first idea of using `log10` was

```python
# python
abs(round(math.log10(i)))
```

but more correctly this should be

```python
# python
math.floor(-math.log10(i))
```

$$
\text{digits of precision} \approx −\log_{10}(\text{error})
$$

for the above

### Example:

```
i = 0.00351
log10 ≈ -2.4547
```

- round(-2.4547) → -2 → digits = 2 ❌
- correct: digits ≈ 3 ✅

👉 Rounding loses precision near boundaries

## 🧭 Summary

Estimating decimal precision from magnitude of a number

Difference:

Method            | Behavior
------------------|------------------------
log10 + round     | ❌ unstable near boundaries
Your loop         | ✅ intuitive but step-based
floor(-log10(x))  | ✅ mathematically correct

````python
import math

stop = 0.01
step = 0.0005
start = 0.00001
# Calculate the number of steps required
num_steps = int((stop - start) / step)

def digits_of_precision(value):
    digits = 0
    while round(value*(10 ** digits)) < 1.0:
        digits += 1
    return digits
    
decimal_range = [start + i * step for i in range(num_steps)]
for i in decimal_range:
    if i == 0.0:
        continue
    log10 = math.log10(i)
    log10_digits = abs(round(math.log10(i)))
    d_of_prec_val = digits_of_precision(i)
    correct = math.floor(-math.log10(i))
    message = f"number:{i:.5f}, log10:{log10:.4f}, digits:{log10_digits}, digits of precision:{d_of_prec_val}, correct?:{correct}"
    print(message)
````
